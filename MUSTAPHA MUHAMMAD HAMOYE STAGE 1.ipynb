```python
#Which feature has missing values and what is the total number of missing value and percentage of the missing rows as a factor of the total number of rows in three decimal places? 
#(Enter answer in the format: Feature: xxx, Total: xxx, Percent: xxx)
#Feature: fuel_unit, Total: 180, Percent: 0.610

fuel_data_df.isnull().sum() #to view the column with missing values which is "fuel_unit"

Missing_row =  fuel_data_df["fuel_unit"].isnull().sum() #to get the number of missing row in fuel_unit column which is 180

fuel_data_df.shape # to display the number of rows and columns in the dataset

Total_row = fuel_data_df.shape[0] # to dispay only the number of rows, which is the total number of rows which is 29523

percentage = (Missing_row / Total_row) * 100 # formula for percentage

print (percentage) # to print the answer
```

```python
#Which of the following fuel type code has the lowest average fuel cost per unit burned?
fuel_type_lowest = fuel_data_df[["fuel_type_code_pudl","fuel_cost_per_unit_burned"]].groupby("fuel_type_code_pudl").mean()
#to form a dataframe of fuel type and and avearge fuel cost per unit burned using **groupby** and **.mean()**
fuel_type_lowest.sort_values("fuel_cost_per_unit_burned") #to sort the values in ascending order view the lowest value of fuel_cost_per_unit_burned
#gas = 13.659397
```

```python
#fuel type code  that has lowest average fuel cost per unit burned is gas
```

```python
#What is the standard deviation and 75th percentile of the measure of energy per unit (Fuel_mmbtu_per_unit) in two decimal places?
#Answer = 10.60 and 17.01
fuel_data_df.describe() #using view all statistical information on the dataset which includes the fuel_mmbtu_per_unit column

A = fuel_data_df.describe() # assigning variable "A" to the describe syntax
print (A.loc["75%","fuel_mmbtu_per_unit"]) #to print the 75th percentile using the slice method

print (A.loc["std","fuel_mmbtu_per_unit"]) #to print the standard deviation using the slice method

#standard deviation = 10.60
#75th percentile = 17.01
```

```python
#Which year has the highest average fuel cost per unit delivered?
fuel_year_highest = fuel_data_df[["report_year","fuel_cost_per_unit_delivered"]].groupby("report_year").mean()
#to form a dataframe of report year and avearge fuel cost per unit delivered using **groupby** and **.mean()**
fuel_year_highest.sort_values("fuel_cost_per_unit_delivered",ascending=False) #to sort the values in descending order view the lowest value
#1997
```

```python
##The year that has the highest average fuel cost per unit delivered  1997
```

```python
#What is the skewness and kurtosis for the fuel quantity burned in two decimal places?
skewness = fuel_data_df["fuel_qty_burned"].skew()# to extract quantity of fuel burned from the dataset and find the skewness
print (skewness)

Kurtosis = fuel_data_df["fuel_qty_burned"].kurtosis()#to extract quantity of fuel burned and find the kurtosis
print (Kurtosis)
```

```python
#skewness and kurtosis for the fuel quantity burned is 15.85 and 651.37
```

```python
#For the fuel type coal, what is the percentage change in the fuel cost per unit burned in 1998 compared to 1994?
condition = fuel_data_df[fuel_data_df["fuel_type_code_pudl"] == "coal"]
#setting the fuel_data_df to condition to display only the dataset  that includes fuel type coal

coal_condition_1994 = condition[condition["report_year"] == 1994]
coal_condition_1998 = condition[condition["report_year"] == 1998]
#setting the variable "condition" to another condition to display only the data that was report in 1994 and 1998 

total_coal_94 = coal_condition_1994["fuel_cost_per_unit_burned"].sum(axis=0)
total_coal_98 = coal_condition_1998["fuel_cost_per_unit_burned"].sum(axis=0)
# to display the fuel cost per unit burned in 1994 and 1998


print ((total_coal_98 - total_coal_94)/total_coal_94 * 100)
#to find the percentage change in fuel cost per unit burned in 1998 and 1994 which is approximately -21% 
```
